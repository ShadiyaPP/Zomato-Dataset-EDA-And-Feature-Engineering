```python
#1. missing values 
df.isnull().sum()
```

```python
#null value column selection
[features for features in df.columns if df[features].isnull().sum()>0] 
```

```python
#visualizating the null value 
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#importing the data about Country
df_county = pd.read_excel('Country-Code.xlsx')
df_county.head()
```

```python
#Merging the two dataframe
final_df = pd.merge(df,df_county,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
#datatypes
final_df.dtypes
```

```python
#Value count in Country column
final_df.Country.value_counts()
```

```python
#selecting the Country name and values for ploting pie chart
country_name = final_df.Country.value_counts().index
```

```python
#values
values = final_df.Country.value_counts().values
```

```python
#create a pie chart
plt.pie(values,labels=country_name)
```

```python
#create a pie chart of top 3
plt.pie(values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation : Zomato maximum records or transactions are from India, after that USA and UK.

```python
#Perform groupby operation on Aggregate rating,ating color,Rating text
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
rating = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

Observations

1. when the rating is 4.5 to 4.9 --------------> Excelent
2. when the rating is 4.0 to 4.4 --------------> Very good
3. when the rating is 3.5 to 3.9 --------------> Good
4. when the rating is 2.5 to 3.4 --------------> Average	
5. when the rating is 1.8 to 2.4 --------------> Poor


```python
#bar plot on Aggregate rating and Rating count
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating',y ='Rating count',data = rating )
```

```python
sns.barplot(x = 'Aggregate rating',y ='Rating count',hue='Rating color',data = rating,palette=['blue','red','orange','yellow','green','green'] )
```

Observations: 
1. Not rated count is very high
2. Maximum no of ratings are between 2.5 to 4.4

```python
#count plot
sns.countplot(x='Rating color',data = rating,palette=['blue','red','orange','yellow','green','green'])
```
